```python
import pandas as pd
data=pd.DataFrame({'value':[1,2,3,45,2,13,98]})
mean=np.mean(data['value'])
std=np.std(data['value'])
print("mean:",mean)
print("std:",std)
thersold=2
outliers=[]
for i in data['value']:
    z=(i-mean)/std
    if z>thersold:
        outliers.append(i)
print("outliers:",outliers)       
```

```python
q1=data['value'].quantile(0.25)
q3=data['value'].quantile(0.75)
IQR=q3-q1
lowerbound=q1-1.5*IQR
upperbound=q3+1.5*IQR
outliers=data[(data['value']<lowerbound) | (data['value']>upperbound)]
print(outliers)
```

```python
m=np.mean(data['value'])
print("mean:",m)
for i in data['value']:
    if i <lowerbound or i > upperbound:
        data['value']=data['value'].replace(i,0)
```

```python
from sklearn.datasets import load_iris
df=load_iris()
df
```

```python
x=df.data
y=df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,accuracy_score
re2=r2_score(y_test,y_pred)
print(re2)
```
